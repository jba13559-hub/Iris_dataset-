<h4>SCATTER PLOT:





It displays relationship between two numerical features of data by using points.

**Purpose:**

Help identify correlations or trends btw variables

Detect outliers and shows clusters in data.

```python
# Set figuresize of plot with widthsize 8inch and height is 6inch 
plt.figure(figsize = (8,6))

# Scatter plot btw Sepal Length on x_axis and Sepal Width on y_axis with different color and style of species exist.
sns.scatterplot(df,x='sepal-length',
                y='sepal-width',
                hue='species',
                style='species',
                s=70,      # point size is 70
                alpha=0.9   # alpha is transparency of points
                )
plt.title('SCATTER PLOT BETWEEN SEPAL-LENGTH AND SEPAL-WIDTH')
plt.show()
```

**INSIGHTS:**  This scatter plot presents a much more crowded and mixed distribution, showing that sepal dimensions are far less distinct between species. While Iris-setosa  tends to separate itself slightly but the data points for Iris-versicolor and Iris-virginica overlap heavily in the center of the plot. This significant overlap indicates that sepal length and width do not vary cleanly enough between versicolor and virginica to tell them

```python
# Set the figure size and creating scatterplot between Petal Length and Petal width
plt.figure(figsize=(8,6))
sns.scatterplot(df,x='petal-length',y='petal-width',
                hue='species',
                style='species',
                s=70)
plt.title('Scatter Plot Btw Petal-Length and Petal Width')
plt.show()
```

**INSIGHTS:**  This graph shows that the Iris-setosa species forms a completely isolated cluster in the lower-left corner, revealing that its petals are uniquely small in both length and width. Meanwhile, Iris-versicolor and Iris-virginica form elongated clusters that show a strong positive linear correlation, meaning that as petal length increases, petal width increases proportionally. Although there is a slight overlap between versicolor and virginica, their clear separation demonstrates that petal measurements are highly effective features for classifying these species

<h5>Now there are another four graphs to discover more distinct clusters and provide valuable cross dimensional insights.
 

sepal-length vs petal-length

sepal-length vs petal-width

sepal-width vs petal-length

sepal-width vs petal-width

All these graphs will arrange as two on top  row and two on bottom  row for analysis and better comparison.

```python


# Create figure with width=14 and height=10 inches
plt.figure(figsize=(14,10))


# create first subplot in 2x2 grid
plt.subplot(2,2,1)
sns.scatterplot(df,x='petal-length',y='sepal-length',hue='species',style='species')
plt.title('Scatter plot btw Sepal length and petal length')

# Create 2nd subplot in 2x2 grid
plt.subplot(2,2,2)
sns.scatterplot(df,x='sepal-length',y='petal-width',hue='species')
plt.title('RELATIONSGHIP BTW SEPAL-LENGTH AND PETAL-WIDTH')


# Create 3rd subplot in 2x2 grid
plt.subplot(2,2,3)
sns.scatterplot(df,x='sepal-width',y='petal-length',hue='species')
plt.title('SCATTERPLOT BTW SEPAL-WIDTH AND PETAL-LENGTH')

# Create fourth subplot in 2x2 grid
plt.subplot(2,2,4)
sns.scatterplot(df,x='sepal-width',y='petal-width',hue='species')
plt.title('SCATTERPLOT BTW SEPAL-WIDTH AND PETAL-WIDTH')

#plt.tight_layout()     # It automatically adjust spacing btw plots and titles
plt.subplots_adjust(hspace=0.5)   # hspace increase gap btw upper and lower row ,prevent overlapping
plt.show()
```

**INSIGHTS:**  Cross-dimensional scatter plots reveal that pairing a petal feature with a sepal feature creates highly distinct, slanted clusters that map out clear diagonal boundaries for classification. These combinations expose a strong, proportional growth relationship across dimensions, proving that as the inner petal scales up, the outer sepal expands along with it. Ultimately, these mixed plots show that even when messy sepal data is involved, the dominant, sharp signal from the petal features successfully pulls the species apart into predictable groups.

<h4> Histplot()

It shows the distribution of data in a numerical variable by grouping values into bins.

**Purpose:**

Reveals the frequency distribution of data.

Helps understand data skewness, spread and shape.

```python

# Set figuresize with width=10 and height=6 inches
plt.figure(figsize=(10,6))
# Create subplot in 1x2 grid with index= 1
plt.subplot(1,2,1)
sns.histplot(df,x='sepal-length',hue='species',kde=True,bins=15)
plt.title('Sepal Length Data Distribution')


# Create subplot in 1x2 grid with index= 2
plt.subplot(1,2,2)
sns.histplot(df,x='sepal-width',hue='species',kde=True,bins=15)
plt.title('Sepal Width Data Distribution')
plt.tight_layout()

plt.show()
```

**INSIGHTS:** The sepal histograms display three heavily blended, overlapping color distributions that stack together to form a single,  bell-shaped curve. This visual blending proves these features fail to separate the species effectively.

```python

# Create figuresize with width=12 and height=6 inches
plt.figure(figsize=(12,5))
# Create subplot in 1x2 grid with index 1
plt.subplot(1,2,1)
sns.histplot(df,bins=15,x='petal-length',hue='species',kde=True)    # bins mean number of bars in plot
plt.title('Petal Length Data Distribution')


# Create subplot in 1x2 grid with index 2
plt.subplot(1,2,2)
sns.histplot(df,x='petal-width',hue='species',kde=True,bins=15)   # kde = True mean curve is shown in plot
plt.title('Petal Width Data Distribution')


plt.show()
```

**INSIGHTS:** For petal length and width, the plot shows a completely isolated, solid block of color for Setosa on the left, separated by a wide, empty gap from the Versicolor and Virginica curves, which sit on the right with only minor overlapping boundaries.The colored peaks for petal dimensions act as sharp, distinct signatures for clean classification.

<h4> Boxplot()

A box plot summarizes data distributon using quartiles, median, outliers. It detect ouliers and compare distributions across groups.Boxplot usually have vertical lines extending from the boxes called whiskers.. so it is also named as **Box-and-Whisker Plot**.

**Purpose:**

Shows data spread and central tendency.

Identifies potential outliers quickly.

```python
plt.figure(figsize=(12,6))
plt.subplot(1,2,1)
sns.boxplot(df,x='sepal-length',y='species',hue='species')
plt.title('BOX PLOT FOR SEPAL LENGTH')


plt.subplot(1,2,2)
sns.boxplot(df,x='sepal-width',y='species',hue='species')
plt.title('BOX PLOT FOR SEPAL WIDTH')

plt.tight_layout()
plt.show()
```

**INSIGHTS:** This reveals a heavy distribution overlap between species , with sepal width uniquely containing a few prominent outliers. This lack of distinct boundaries shows that sepal measurements are highly crowded, making them less reliable for individual species classification.

```python

# Set figure size of width=12 and height=6 inches
plt.figure(figsize=(10,5))

# Set a subplot in 1x2 grid as the index of plot is 1
plt.subplot(1,2,1)
sns.boxplot(df,x='petal-length',y='species', hue='species')
plt.title('BOX PLOT FOR PETAL LENGTH')

# Set a subplot in 1x2 grid as the index of plot is 2
plt.subplot(1,2,2)
sns.boxplot(df,x='petal-width',y='species',hue='species')
plt.title('BOX PLOT FOR PETAL WIDTH')

plt.tight_layout()    # It automatically adjust spacings btw plots and titles
plt.show()
```

**INSIGHTS:** This displays clear-cut gaps between the species, showing Setosa is completely isolated and minimal overlap elsewhere. This sharp separation demonstrates that petal measurements possess strong predictive power that is signal for better classification

<h4> Barplot()

A barplot shows the average value of a numerical column for each category of categorical column. It represents values of different categories using rectangular bars.. 

**Purpose:**

Compare numerical values across categories.

Makes differences btw groups easy to visualize.

<h5> For better visualization and comparison, i will arrange graphs as two on the top row and two on bottom row

```python
plt.figure(figsize=(14,8))
plt.subplot(2,2,1)
sns.barplot(df,y='petal-length',x='species',hue='species')
plt.title('Bar Plot for Petal Length')

plt.subplot(2,2,2)
sns.barplot(df,x='species',y='petal-width',hue='species')
plt.title('Bar Plot for Petal Width')

plt.subplot(2,2,3)
sns.barplot(df,x='species',y='sepal-length',hue='species')
plt.title('Bar Plot for Sepal Length')

plt.subplot(2,2,4)
sns.barplot(df,x='species',y='sepal-width',hue='species')
plt.title('Bar Plot for Sepal Width')
plt.tight_layout()
plt.subplots_adjust(hspace=0.5)
plt.show()
```

**INSIGHTS:** The above row clearly shows a stepwise increase in average petal length and width from Setosa to Versicolor to Virginica, marking them as the most distinguishing features.
While on seeing 2nd row,it clearly seen sepal lengths show a much more gradual growth across species, while Setosa uniquely holds the highest average sepal width despite having the smallest dimensions everywhere else.
